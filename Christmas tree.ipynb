```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
t1 = np.linspace(3.1*np.pi,3.9*np.pi,200) #樹葉
r1 = np.cos(2*t1)
x1 = r1*np.cos(t1)
y1 = r1*np.sin(t1)

for i in range(1,10,2): #有幾層樹葉，每層增加兩個
    for j in range(i):
        plt.plot(x1 - (i//2) + j, y1, c='#00A800', lw=3) #樹葉
    x = np.linspace(-0.7-(i//2), 0.7+(i//2), 200) #紅色彩帶
    y = ((np.sin(5*x))/6) - 1.4*(i//2) #紅色彩帶
    plt.plot(x, y, c='#E60000', lw=2, alpha=0.5) #紅色彩帶
    y1 = y1 - 1.4

t2 = np.linspace(0,2*np.pi,200) #樹幹
r2 = (np.cos(2*t2)+1)/2
x2 = r2*np.cos(t2)
y2 = r2*np.sin(t2)

for i in range(5): #有幾層樹幹
    plt.plot(x2, y2-6.2-(i/3), c='#805300', lw=4)

t3 = np.linspace(0,np.pi,200) #黃色五角星星
r3 = np.sin(5*t3)
plt.plot(r3*np.cos(t3), r3*np.sin(t3)+1.5, c='#FFFF00')    
    
ax = plt.gca()
ax.set_aspect('equal')
```
