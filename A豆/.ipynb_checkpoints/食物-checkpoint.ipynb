```python
def 菜單中(message):
    for i in 餐點.keys():
        if i in message:
            return i
```

```python
def 食物(name):
    print('%s: %s' %('Aっ豆', rn.choice(建議)))
    message = input('%s:' %(name))
    while not any(i in message for i in 決定):
        if 菜單中(message):
            print('%s: %s 可以參考 %s,如何呢？' %('Aっ豆', name, rn.choice(餐點[菜單中(message)])))
        else:
            print('%s: %s 你可以參考 %s,如何呢？' %('Aっ豆', name, rn.choice(餐點[rn.choice(list(餐點.keys()))])))
        message = input('%s:' %(name))
    print('%s: %s' %('Aっ豆', rn.choice(回決定)))
```
