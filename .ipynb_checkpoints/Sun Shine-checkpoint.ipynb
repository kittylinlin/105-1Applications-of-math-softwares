## 畫張太陽臉吧！

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
#太陽臉
r = 12
t = np.linspace(0,2*np.pi,200)
x = r*np.cos(t)
y = r*np.sin(t)
plt.plot(x,y,'y',lw='3')

r1 = 16 #光芒中心離原點的距離
for i in np.linspace(0,2*np.pi,12): #光芒數
    r2 = 3 #光芒大小
    t1 = np.linspace(0,2*np.pi,4) #光芒的形狀
    x1 = r1*np.cos(i)+(r2*np.cos(t1)*np.cos(i)-r2*np.sin(t1)*np.sin(i)) #光芒在原點畫出後，先依光芒數旋轉，在平移到光芒距原點距離
    y1 = r1*np.sin(i)+(r2*np.sin(t1)*np.cos(i)+r2*np.cos(t1)*np.sin(i))
    plt.plot(x1,y1,'y',lw='3')

r3 = 2 #眼睛大小
t3 = np.linspace(0,2*np.pi,20)

x3 = -5+r3*np.cos(t3) #左眼
y3 = 4+r3*np.sin(t3)

x4 = 5+r3*np.cos(t3) #右眼
y4 = 4+r3*np.sin(t3)

x5 = 0+2*np.cos(t3) #嘴巴
y5 = -4+4*np.sin(t3)

plt.plot(-4,4,'ok',6,4,'ok') #眼珠
plt.plot(x3,y3,'k',x4,y4,'k') #眼睛
plt.plot(x5,y5,'r',lw='2') #嘴巴
    
ax = plt.gca()
ax.set_aspect('equal')
```
