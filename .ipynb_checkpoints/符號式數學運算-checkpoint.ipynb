```python
import random as rn
from IPython.display import display, HTML
```

```python
def makequest():
    n = rn.randint(1,5)
    f = rn.randint(0,10)
    for i in range(1,n):
        f = f + rn.randint(0,10)*x**i
    return f,n
```

```python
display(HTML('<p>來玩微分吧！請對以下多項式進行微分</p>'))
q = makequest()
display(HTML('<p>f(x) = $'+ latex(q[0]) +'$</p>'))
ans = input('請以降冪方式輸入係數，並以逗號分開：')
ans = ans.split(',')
answer = diff(q[0])
wrong = 0
for i in range(q[1]-1):
    if(int(ans[i]) != int(answer.coeff(x,n=(q[1]-2-i)))):
        wrong = 1
        break
if(wrong == 1):
    print('答錯囉！！')
else:
    print('答對哩～～')
```

```python
def findstr():
    n = int(input('please enter digits of pi:'))
    st = str(pi.n(n))
    ans = input('please input your birthday:')
    if(ans in st):
        print('%s is at the index %s' %(ans, st.find(ans)))
    else:
        print('sorry, try more digits of pi')
```
